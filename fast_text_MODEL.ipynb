```python
  import tensorflow as tf
  from tensorflow.keras.preprocessing.text import Tokenizer
  from tensorflow.keras.preprocessing.sequence import pad_sequences
  from sklearn.model_selection import train_test_split
```

```python
tokenizer = Tokenizer(
    num_words=None,
    filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
    lower=True,
    split=' ',
    char_level=False,
    oov_token=None,
    analyzer=None,
)
tokenizer.fit_on_texts(df["comments"])
```

```python
x = tokenizer.texts_to_sequences(df["comments"])
```

```python
x=pad_sequences(x,padding="post",truncating='post',maxlen=70)


#x = pad_sequences(x)
```

```python
# Define regex pattern
pattern = r'[^\w\s]'

# Apply regex pattern to text column
df["comments"] = df['comments'].str.replace(pattern, '')


label_dict = {}
unique_labels = df['labels'].unique()
for index, label in enumerate(unique_labels):
    label_dict[label] = index

df['numerical_label'] = df['labels'].map(label_dict)
```

```python
df['labels'] = '__label__' + df['labels'].astype(str)
df['comments_label'] = df['labels'] + ' ' + df['comments']
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.3)
```

```python
train.shape, test.shape
```

```python
train.to_csv("comments.train", columns=["comments_label"], index=False, header=False)
test.to_csv("comments.test", columns=["comments_label"], index=False, header=False)
```

```python
import fasttext

model = fasttext.train_supervised(input="/content/comments.train",autotuneValidationFile='/content/comments.test', autotuneDuration=600,lr=1.0, epoch=100, wordNgrams=3, bucket=200000, dim=50, loss='ova')
```

```python
model.test("/content/comments.test")
```

```python
model_update = fasttext.train_supervised(input="/content/comments.train",autotuneValidationFile='/content/comments.test',autotuneDuration=600, lr=1.0,dim=100, epoch=100, wordNgrams=2)
```

```python
model_update.test("/content/comments.test")
```
